```python
"""Write both recursive and iterative function to compute the factorial of a number.
How does the performance of the recursive function compare to that of an iterative version?
   F(0)=F(1) = 1
   F(n) = F(n-1)*n, for n >=2"""

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(6))#ans is 720
```

```python
Write both  recursive and iterative function to compute the Fibonacci sequence. 
        How does the performance of the recursive function compare to that of an iterative version?
           F(1)=F(2) = 1
   F(n) = F(n-1)+F(n-2), for n >= 3
```

```python
#Recursive version
def fib_recursive(n):
    if n <=0:
        return 0
    elif n<=2:
        return n
    else:
        return fib_recursive(n-1)+fib_recursive(n-2)
    
#Iterative version
def fib_iterative(n):
    if n == 0 or n == 1:
        return 1
    n_Sum = 0
    n_1_Sum=1
    n_2_Sum=1
    for i in range(2,n+1):
        n_Sum = n_1_Sum + n_2_Sum
        n_2_Sum = n_1_Sum
        n_1_Sum = n_Sum
    return n_Sum

print(fib_recursive(9))# Ans is 55
print(fib_iterative(9))# Ans is 55


# Performance comparison

# Recursive version is lot slower than iterative version

# In recursive version we calculate fibonacci value of all numbers less than n-2 twice
# Once during calculation of f(n)
# And during calculation of f(n-1)


```

```python
import turtle

def tree(branchLen,t):
    if branchLen > 5:
        t.forward(branchLen)
        t.right(20)
        tree(branchLen-15,t)
        t.left(40)
        tree(branchLen-15,t)
        t.right(20)
        t.backward(branchLen)

def main():
    t = turtle.Turtle()
    myWin = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.color("green")
    tree(75,t)
    myWin.exitonclick()

main()
```


Modify the recursive tree program to produce a more realistic looking tree with various brnach length/thickness, and braching angles.

```python
"""Modify the thickness of the branches so that as the branchLen gets smaller, the line gets thinner.
Modify the color of the branches so that as the branchLen gets very short it is colored like a leaf."""
import turtle

def tree(branchLen,t):
    if branchLen > 5:
        t.forward(branchLen)
        t.width(branchLen/25) # Modify the thickness of the branches so that as the branchLen gets smaller, the line gets thinner
        t.right(20)
        t.width(branchLen / 25)
        tree(branchLen-15,t)
        t.left(40)
        t.width(branchLen / 25)
        tree(branchLen-15,t)
        t.right(20)
        t.width(branchLen / 25)
        t.backward(branchLen)
        if branchLen < 11: #  Modify the color of the branches so that as the branchLen gets very short it is colored like a leaf
            t.color("blue")
        else:
            t.color("green")

def main():
    t = turtle.Turtle()
    myWin = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.color("green")
    tree(100,t)
    myWin.exitonclick()

main()
```

```python
import turtle
def draw_sierpinski(length,depth):
    if depth==0:
        for i in range(0,3):
            t.forward(length)
            t.left(120)
    else:
        draw_sierpinski(length/2,depth-1)
        t.forward(length/2)
        draw_sierpinski(length/2,depth-1)
        t.backward(length/2)
        t.left(60)
        t.forward(length/2)
        t.right(60)
        draw_sierpinski(length/2,depth-1)
        t.left(60)
        t.backward(length/2)
        t.right(60)

window = turtle.Screen()
t = turtle.Turtle()
draw_sierpinski(100,2)
window.exitonclick()
```
