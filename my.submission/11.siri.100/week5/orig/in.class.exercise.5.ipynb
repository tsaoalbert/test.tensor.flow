```python
def hash(astring, tablesize):
    sum = 0
    for pos in range (len(astring)):
        sum = sum + ord(astring[pos]) + pos
    return sum%tablesize

print(hash("test string",10))
```

```python
#Bubble sort
def foo(alist, i, j):
    print (i, end=": ")
    for k in range(0, len(alist) ):
        s = "{0:3}".format(alist[k])
        if k==i:
            s += "*"
        elif k==j:
            s += "-"
        else:
            s += " "
        print (s, end="")
    print ()
def bubbleSort(L):
    
    for k in range(len(L)-1,0,-1):
        foo (L,k,k)
        for i in range(k):
            if L[i]>L[i+1]:
                L[i],L[i+1]=L[i+1],L[i]

alist = [7, 2,9, 3,4,10, 5,1, 6,8]
bubbleSort(alist)
assert(alist==sorted(alist))
```

```python
#Selection Sort

def selectionSort(alist):
  for i in range(len(alist)-1,0,-1):
    j=i
    for k in range(0,i):
      if alist[k]>alist[j]:
        j = k 
    foo (alist, i, j)
    alist[i], alist[j] = alist[j], alist[i]

alist = [7, 2,9, 3,4,10, 5,1, 6,8]
selectionSort(alist)
assert(alist==sorted(alist))
```

```python
#Insertion Sort

def insertionSort(alist):
  foo(alist, 0,0)
  for i in range(1,len(alist)):
    x = alist[i]
    j = i
    while j>0 and alist[j-1]>x:
      alist[j]=alist[j-1]
      j -= 1
    alist[j] = x
    foo(alist, i,j)

s =  [7, 2,9, 3,4,10, 5,1, 6,8]
insertionSort(s)
assert ( s==sorted(s))
for i in range (1,len(s)):
    assert s[i-1] <= s[i] 
```

```python
#Merge Sort

def mergeSort(alist, level, first, last, A,B):
  # print("Splitting ",alist)
  ans = level
  if last==first:
    A[level][first] = alist[first]
    A[level+1][first] = alist[first]
  elif last-first>1:
    mid = (last+first)//2
    B[level+1][mid-1] = "**"
    for i in range (first,last):
        A[level+1][i] = alist[i]
    l1 = mergeSort(alist, level+1, first, mid, A, B)
    l2 = mergeSort(alist, level+1, mid, last, A, B )
    ans = max (l1,l2)
    i=first
    j=mid
    k=0
    L = [0]*(last-first)
    while i<mid and j<last:
      if alist[i]<alist[j]:
        L[k]=alist[i]
        i=i+1
      else:
        L[k]=alist[j]
        j=j+1
      k=k+1

    while i<mid:
      L[k]=alist[i]
      i=i+1
      k=k+1

    while j<last:
      L[k]=alist[j]
      j=j+1
      k=k+1
    for i in range (k):
        alist[first+i]=L[i]
        A[level][first+i] = L[i]
  return ans 

def traceSorting (k, n, A,B):
    print ()
    for i in range (k+1) :
        print (i, end=": ")
        for j in range (n):
            x = A[i][j]
            if x==0:
                x=""
            else:
                x = str(x)
            print ( "{0:^2}{1:2}".format( x, B[i][j] ),  end="")
        print ()

def unitTest ():
    x = [54,26,93,17,77,31,44,55,20,37,42,99,106,2001]
    y = [21, 1, 26, 45, 29, 28, 2, 9, 16, 49, 39, 27, 43, 34, 46, 40] 
    x = [7, 2,9, 3,4,10, 5,1, 6,8]

    for z in [x,y]:
        n = len(z)
        A = []
        B = []
        C = [0]*n
        for i in range (n):
            A.append ( C.copy() ); B.append( ["  "]*n)
        level = mergeSort(z, 0, 0, n, A,B)
        assert(z==sorted(z))
        traceSorting (level, n, A,B)

unitTest()
```
