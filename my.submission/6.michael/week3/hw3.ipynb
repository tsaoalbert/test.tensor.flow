* 3.1 (10%) Modify the infix-to-postfix algorithm so that it can handle errors.
* 3.2 (10%) Modify the postfix evaluation algorithm so that it can handle errors.
* 3.3 (10%) Implement a direct infix evaluator that combines both infix-to-postfix conversion 
* 3.4 (10%) Turn your direct infix evaluator from the previous problem into a calculator that can handles arithmetic operators +,-,*,/, //, %, ** and parenthesis.

```python
#3.1 Answer:

class Stack:
     def __init__(self):
         self.items = []
     def isEmpty(self):
         return self.items == []
     def push(self, item):
         self.items.append(item)
     def pop(self):
         return self.items.pop()
     def peek(self):
         return self.items[len(self.items)-1]
     def size(self):
         return len(self.items)

def infixToPostfix(infixexpr):
    prec = {}
    prec["**"] = 4
    prec["//"] = 3
    prec["*"] = 3
    prec["/"] = 3
    prec["%"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1
    opStack = Stack()
    postfixList = []
    tokenList = infixexpr.split()
    charaters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    operators = "+-*/%\(\)"
    inputList = list(tokenList)
    if tokenList.count('(') != tokenList.count(')'):
        return "The input is incorrect."
    while '(' in inputList:
        inputList.remove('(')
    while ')' in inputList:
        inputList.remove(')')
    for i in range(len(inputList)):
        if not (i%2 == 0 and (inputList[i] in charaters or inputList[i].isdigit()) \
                or (i%2 == 1 and (inputList[i] in operators or inputList[i] == "//" or inputList[i] == "**"))):
            return "Input incorrect."     

    for item in inputList:
        if not ((item in charaters and len(item) == 1) or item in operators or item == "//" or item == "**" or item.isdigit()):
            return "Input incorrect." 

    for token in tokenList:
        if token in charaters or token.isdigit():
            postfixList.append(token)
        elif token == '(':
            opStack.push(token)
        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()
        else:
            while (not opStack.isEmpty()) and \
               (prec[opStack.peek()] >= prec[token]):
                  postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.isEmpty():
        postfixList.append(opStack.pop())
    return " ".join(postfixList)

print(infixToPostfix(" ( A - C ) ** E + CF // D"))
print(infixToPostfix(" G * ( A / D ) ** E + C // D"))
print(infixToPostfix("( A + D ) * C - ( D - E ) * ( F + G )"))
```

```python
#3.2 Answer:

def postfixEval(postfixExpr):
    try:
        operandStack = Stack()
        tokenList = postfixExpr.split()
        digitAmount = 0
        operatorAmount = 0
        for token in tokenList:
            if token.isdigit():
                digitAmount += 1
            elif token in "+-*/%" or token == "//" or token == "**":
                operatorAmount += 1
            else:
                return "Input incorrect."
        if digitAmount != (operatorAmount + 1):
            return "Input incorrect."
        for token in tokenList:
            if token.isdigit():
                operandStack.push(int(token))
            else:
                operand2 = operandStack.pop()
                operand1 = operandStack.pop()
                result = doMath(token,operand1,operand2)
                operandStack.push(result)
        return operandStack.pop()
    except NameError:
        return "divided by 0!"
    except:
        return "Unexpected error!"

def doMath(op, op1, op2):
    if op == "*":
        return op1 * op2
    elif op == "/":
        if op2 == 0:
            raise NameError("divided by 0!") 
        return op1 / op2
    elif op == "%":
        if op2 == 0:
            raise NameError("divided by 0!") 
        return op1 % op2
    elif op == "+":
        return op1 + op2
    elif op == "-":
        return op1 - op2
    elif op == "**":
        return op1 ** op2
    elif op == "//":
        if op2 == 0:
            raise NameError("divided by 0!") 
        return op1 // op2

print(postfixEval('9 28 + 3 4 + /'))
```

```python
# 3.3 Answer:
def infixEval(infixExpr):
    return postfixEval(infixToPostfix(infixExpr))

print(infixEval("( 10 + 11 ) * 3 - 4"))
```

```python
class DoubleNode:
    def __init__(self, data):
        self.data = data
        self.front = None
        self.back = None
    
class deque:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def AddFromHead(self, item):
        temp = DoubleNode(item)
        if self.head == None:
            self.head = temp
            self.tail = temp
        else:
            temp.back = self.head
            self.head.front = temp
            self.head = temp

    
    def AddFromTail(self, item):
        temp = DoubleNode(item)
        if self.tail == None:
            self.head = temp
            self.tail = temp
        else:
            temp.front = self.tail
            self.tail.back = temp
            self.tail = temp

    def RemoveFromHead(self):
        if self.head == None:
            return False
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return
        self.head = self.head.back
        self.head.front = None
    
    def RemoveFromTail(self):
        if self.tail == None:
            return False
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return
        self.tail = self.tail.front
        self.tail.back = None

    def __str__(self):
        current = self.head
        if current == None:
            return ""
        else:
            stra = str(current.data) + " "
        while current.back != None:
            current = current.back
            stra += str(current.data) + " "

        return stra

print("This is a deque:")
dequea = deque()
dequea.AddFromHead(11)
print(dequea)
dequea.AddFromHead(22)
print(dequea)
dequea.AddFromTail(33)
print(dequea)
dequea.RemoveFromHead()
print(dequea)
dequea.RemoveFromTail()
print(dequea)
    
class stack(deque):
    def pop(self):
        temp = self.tail
        self.RemoveFromTail()
        return temp
    def push(self, item):
        self.AddFromTail(item)

class queue(deque):
    def enqueue(self,item):
        self.AddFromTail(item)
    def dequeue(self):
        self.RemoveFromHead()

print("This is a stack, pop removes the tail, and push adds from the tail:")
stacka = stack()
stacka.push(11)
print(stacka)
stacka.push(22)
print(stacka)
stacka.pop()
print(stacka)

print("This is a queue, enqueue adds from the tail, and dequque removes from the head:")
queuea = queue()
queuea.enqueue(11)
print(queuea)
queuea.enqueue(22)
print(queuea)
queuea.dequeue()
print(queuea)    
```
