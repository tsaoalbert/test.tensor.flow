* 3.1 (10%) Modify the infix-to-postfix algorithm so that it can handle errors.
* 3.2 (10%) Modify the postfix evaluation algorithm so that it can handle errors.
* 3.3 (10%) Implement a direct infix evaluator that combines both infix-to-postfix conversion 
* 3.4 (10%) Turn your direct infix evaluator from the previous problem into a calculator that can handles arithmetic operators +,-,*,/, //, %, ** and parenthesis.

```python
#3.1 raise exception for unrecognized character
def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1
    opStack = []
    postfixList = []
    tokenList = infixexpr.split()

    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfixList.append(token)
        elif token == '(':
            opStack.append(token)
        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()
        elif token in "+-*/%":
            while len(opStack) != 0 and \
               (prec[opStack[len(opStack)-1]] >= prec[token]):
                  postfixList.append(opStack.pop())
            opStack.append(token)
        else:
            raise Exception("invalid character %s, please check your expression and enter again"% token)         
    while len(opStack) != 0:
        postfixList.append(opStack.pop())
    return " ".join(postfixList)

print("----------infix to postfix-----------")
print(infixToPostfix("A * B + C * D"))
print()
print("----------throw exception to handle error-----------")
print(infixToPostfix("( A @ B ) * C - ( D - E ) * ( F + G )"))
```

```python
#3.2 raise exception for unrecognized character
def postfixEval(postfixExpr):
    operandStack = []
    tokenList = postfixExpr.split()

    for token in tokenList:
        if token in "0123456789":
            operandStack.append(int(token))
        elif token in "+-*/":
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token,operand1,operand2)
            operandStack.append(result)
        else:
            raise Exception("invalid character %s, please check your expression and enter again"% token)
    return operandStack.pop()

def doMath(op, op1, op2):
    if op == "*":
        return op1 * op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    else:
        return op1 - op2

print("----------post evulation '7 8 + 3 2 + /' -----------")
print(postfixEval('7 8 + 3 2 + /'))
print()
print("----------post evulation '7 8 + 3 2 + / 5 6 + -' -----------")
print(postfixEval('3 4 + 1 2 + / 5 6 + -'))
print("----------throw exception to handle error-----------")
print(postfixEval('3 $ + 1 2 + / 5 6 + -'))
```

```python
#3.3
def infixEval(infixExpr):
    postfixExpr = None
    result = 0
    try:
        postfixExpr = infixToPostfix(infixExpr)
    except Exception as e:
        print(e)
        
    try:
        
        result = postfixEval(postfixExpr)
    except Exception as e:
        print(e)
    
    return result

print("-----------------evaluate 1 * 2 + 3 * 4-----------------")
print(infixEval("1 * 2 + 3 * 4"))
print()
print("-----------------evaluate 1 + ( 6 / 3 ) * 4 / 2-----------------")
print(infixEval("1 + ( 6 / 3 ) * 4 / 2"))
print()
print("----------throw exception to handle error-----------")
print(infixEval("1 ! * 2 + 3 * 4"))
```

* 3.12 (10%) Extend the program from Listing 2.15 to handle palindromes with spaces. For example, I PREFER PI is a palindrome that reads the same forward and backward if you ignore the blank characters.

```python
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = None

    def getData(self):
        return self.data

    def __str__(self):
        return str(self.data)

    def getNext(self):
        return self.next

    def setData(self,newdata):
        self.data = newdata


    def setNext(self,newnext):
        self.next = newnext

class Deque:                            # Deque implementation based on python list
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []
    # O(N)                              # Equivalent Python deque methods
    def addFront(self, item):           # appendLeft
        self.items.insert(0,item)

    # O(1)
    def addRear(self, item):            # append
        self.items.append(item)
    # O(N)
    def removeFront(self):              # popLeft
        return self.items.pop(0)
    # O(1)
    def removeRear(self):               # pop
        return self.items.pop()

    def size(self):
        return len(self.items)

def palchecker(aString):
    aString = ''.join(e for e in aString if e.isalpha())
    aString = aString.upper()
    chardeque = Deque()

    for ch in aString:
        chardeque.addRear(ch)

    stillEqual = True

    while chardeque.size() > 1 and stillEqual:
        first = chardeque.removeFront()
        last = chardeque.removeRear()
        if first != last:
            stillEqual = False

    return stillEqual
A = [ "I PREFER PI","A Nut for a Jar of Tuna", "A But Tuba", "As I Pee, Sir, I See Pisa", \
        "A Toyota's a Toyota", "No X in Nixon", "radar",  "I prefer pi", "Race car" ] 
for s in A:
    print(palchecker (s))
```

```python
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self,newdata):
        self.data = newdata

    def setNext(self,newnext):
        self.next = newnext

class UnorderedList:

    def __init__(self):
        self.head = None
        self.count = 0

    def isEmpty(self):
        return self.head == None
    
    def add(self,item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
        self.count+=1
    
    def size(self):
        return self.count
    
    def search(self,item):
        current = self.head
        found = False
        while current != None and not found:
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()

        return found
    
    def remove(self,item):
        cur = self.head
        pre = Node(-1) #fake head
        pre.next = cur
        while cur != None:
            if cur.getData() == item:
                if self.head == cur:
                    self.head = cur.next
                pre.next = cur.next
                cur = pre.next
                self.count-=1
            else:
                pre = cur
                cur = cur.next
        
    def __str__(self):
        result = []
        cur = self.head
        while cur != None:
            result.append(cur.getData())
            cur = cur.next
        return result
    
    #append at the beginning of the linkedlist
    def append(self, item):
        self.add(item)
    def index(self, node):
        idx = 0
        cur = self.head
        pre = Node(-1)
        while cur != None:
            if cur == node:                
                return idx
            else:
                idx += 1
                pre = cur
                cur = cur.next  
        return -1
    #remove this node
    def pop(self, node):
        cur = self.head
        pre = Node(-1)
        while cur != None:
            if cur == node:
                if cur == self.head:
                    self.head = cur.next
                pre.next = cur.next
                break
            else:
                pre = cur
                cur = cur.next
    #insert before this node
    def insert(self, node, item):
        cur = self.head
        pre = Node(-1)
        while cur != None:
            if cur == node:
                if cur == self.head:
                    self.head = cur.next
                tmp = Node(item)
                tmp.next = cur
                pre.next = tmp
                break
            else:
                pre = cur
                cur = cur.next
                
singleLinkedList = UnorderedList()
singleLinkedList.add(1)
singleLinkedList.add(2)
singleLinkedList.add(1)
singleLinkedList.add(3)
singleLinkedList.add(1)
singleLinkedList.add(4)
singleLinkedList.add(5)
singleLinkedList.add(1)

print("---------Initial linkedlist--------")
print(singleLinkedList.__str__())
print()
print("---------Test removing all 1 from list (remove duplicates)---------")
singleLinkedList.remove(1)
print(singleLinkedList.__str__())
print()
print("---------Test removing nonexisting item 100 from list---------")
singleLinkedList.remove(100)
print(singleLinkedList.__str__())
print()
print("---------print linkedList length---------")
print(singleLinkedList.size())
print()
print("---------Test linkedList toString method---------")
print(singleLinkedList.__str__())
print()
print("---------Test linkedList Append method---------")
singleLinkedList.append(66)
print(singleLinkedList.__str__())
print()
print("---------Test linkedList Pop method---------")
print("Pop the 3rd element from list head")
node = singleLinkedList.head.next.next
singleLinkedList.pop(node)
print(singleLinkedList.__str__())
print()
print("---------Test linkedList Index method---------")
print("find the index of 4th node from list head")
node = singleLinkedList.head.next.next.next
print(singleLinkedList.index(node))
print("---------Test linkedList Insert method---------")
print("insert a new list node w/ value = 888 after list node 5")
node = singleLinkedList.head.next.next
singleLinkedList.insert(node, 888)
print(singleLinkedList.__str__())
print()
```

```python
class Deque:
    def __init__(self, doubleLinkedList):
        self.doubleLinkedList = doubleLinkedList
    
    def addFront(self, item):
        self.doubleLinkedList.addFront(item)
    def addRear (self, item):
        self.doubleLinkedList.addRear(item)
    def removeFront (self): 
        self.doubleLinkedList.removeFront()
    def removeRear(self):
        self.doubleLinkedList.removeRear()
    def isEmpty(self):
        return self.doubleLinkedList.isEmpty()
    def size(self):
        return self.doubleLinkedList.length()
    
    def __str__(self):
        return str(self.doubleLinkedList.toList())
doubleLinkedList = DoublyLinkedList();
deque = Deque(doubleLinkedList)
print("----------initially, an empty Deque----------")
print(deque.__str__())
print()
deque.addFront(1)
print("----------Enqueue 1 from front----------")
print(deque.__str__())
print()
print("----------Enqueue 2 from front----------")
deque.addFront(2)
print(deque.__str__())
print()
print("----------Enqueue 3 from back-----------")
deque.addRear(3)
print(deque.__str__())
print()
print("----------Enqueue 4 from back-----------")
deque.addRear(4)
print(deque.__str__())
print()
print("----------Deque from front-----------")
deque.removeFront()
print(deque.__str__())
print()
print("----------Deque from back-----------")
deque.removeRear()
print(deque.__str__())
```

```python
class Stack:
    def __init__(self, deque):
        self.deque = deque    
    def push(self, item):
        self.deque.addFront(item)
    def pop (self):
        if deque.isEmpty():
            print("stack is already empty")
        else:
            self.deque.removeFront()
    def size(self): 
        return deque.size()
    def isEmpty(self):
        return deque.isEmpty()
    def __str__(self):
        return deque.__str__()
    
doubleLinkedList = DoublyLinkedList();
deque = Deque(doubleLinkedList)
stack = Stack(deque)

print("----------initially, an empty Stack----------")
print(stack.__str__())
print()
print("----------push 1 into stack----------")
stack.push(1)
print(stack.__str__())
print()
print("----------push 2 into stack----------")
stack.push(2)
print(stack.__str__())
print()
print("----------push 3 into stack----------")
stack.push(3)
print(stack.__str__())
print()
print("----------pop stack-----------")
stack.pop()
print(stack.__str__())
print()
print("----------pop stack-----------")
stack.pop()
print(stack.__str__())
print()
print("----------push 4 into stack----------")
stack.push(4)
print(stack.__str__())
print()
print("----------check stack size-----------")
print(stack.size())
```
