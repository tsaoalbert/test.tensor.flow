```python
Answer: 

There will be high chance of collison with hash function above, because two characters in a String can differ 
as litter as 1 in terms of their Unicode values: 'b'-'a' = 1. 
In addition, all anagrams will surely fall into the same bucket(they will be on the same chain)

To magnify this difference and to make the funtion working better on anagrams, I come up w/ hash funcstions below.
There is a chance that the sum variable could easily overflow, we need a long type variable rather than int type to
hold the value of sum.

def hash(word, tableSize):
    sum = 0
    mul = 1
    for index in range(len(word)):
        sum += ord(word[index])*mul
        mul*=10
    return sum%tableSize;
```

```python
Answer: 

When the input list is sorted, picking up middle item is better than choosing the first item in the list.
This is because choosing middle item on sorted list leads to balanced partitions.
Conversely, choosing the first item can lead to heavilhy-skewed (highly unbalanced)subset during divide stage, 
therefore driving down the efficiency.


However I would prefer choosing random item as pivot, because I believe it would achieve efficiency on arbitary list,
regardless of whether it's a sorted list or not.
```

```python
#Improve the time complexity of thefollowing algorithm to be linear time O(N)
def findRepeated(L):
  """
    determines whether all elements in a given list L are distinct
  """
  n=len(L)
  for i in range(n):
    for j in range(i+1, n):
      if L[i]==L[j]:
        return True
  return False
```
