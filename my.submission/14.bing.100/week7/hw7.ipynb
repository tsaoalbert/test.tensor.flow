```python
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

class Vertex:
    def __init__(self,key):
        self.id = key
        self.connectedTo = {}
        self.color = 'white'
        self.dist = sys.maxsize
        self.pred = None
        self.disc = 0    # discover time
        self.fin = 0     # finish   time
        self.neighbor = []
        self.visited = False

    def __str__(self):
        return self.id
    
    def addNeighbor(self,nbr,weight=0):
        self.connectedTo[nbr] = weight

    def setColor(self,color):
        self.color = color

    def setDistance(self,d):
        self.dist = d

    def setPred(self,p):
        self.pred = p

    def setDiscovery(self,dtime):
        self.disc = dtime
        
    def setFinish(self,ftime):
        self.fin = ftime
        
    def getFinish(self):
        return self.fin
        
    def getDiscovery(self):
        return self.disc
        
    def getPred(self):
        return self.pred
        
    def getDistance(self):
        return self.dist
        
    def getColor(self):
        return self.color

    def getConnections(self):
        return self.connectedTo.keys()

    def getWeight(self,nbr):
        return self.connectedTo[nbr]

    def __str__(self):
        return str(self.id) + ' connectedTo: ' + str([x.id for x in self.connectedTo])

    def getId(self):
        return self.id
    
    def addNeighbor(self, vertex):
        self.neighbor.append(vertex)
        
    def getNeighbor(self):
        return self.neighbor

def bfs(start):
    q = Queue()
    hashSet = set()
    q.enqueue(start)
    hashSet.add(start.id)
    res = {}
    dist = 0
    while (q.size() > 0):
        currentVert = q.dequeue()
        dist +=1
        for curr in currentVert.getNeighbor():
            if (curr.id not in hashSet):
                hashSet.add(curr.id)
                q.enqueue(curr)
                if str(start.id) + "-" + str(currentVert.id) not in res:
#                     res[str(start.id) + "-" + str(currentVert.id)] = 0
                    res[str(start.id) + "-" + str(curr.id)] = 1
                else:
                    res[str(start.id) + "-" + str(curr.id)] = res[str(start.id) + "-" + str(currentVert.id)] +1

    return res


def allPairShortestPath(vertices):
    res = []
    for vertex in vertices:
        res.append(bfs(vertex))
    
    return res

v1 = Vertex(1)
v2 = Vertex(2)
v3 = Vertex(3)
v4 = Vertex(4)
v5 = Vertex(5)
v6 = Vertex(6)
v7 = Vertex(7)
v8 = Vertex(8)
v9 = Vertex(9)
v10 = Vertex(10)

v1.addNeighbor(v2)
v1.addNeighbor(v3)
v2.addNeighbor(v3)
v2.addNeighbor(v5)
v3.addNeighbor(v4)
v4.addNeighbor(v6)
v6.addNeighbor(v7)
v7.addNeighbor(v8)
v7.addNeighbor(v9)
v9.addNeighbor(v10)
v5.addNeighbor(v9)
v8.addNeighbor(v5)
v9.addNeighbor(v2)
vertices = [v1,v2,v3,v4,v5,v6,v7,v8,v9,v10]

print("---------print shortest path for each pair: ('from-to': the length of the shortest path)--------")
allPairShortestPath(vertices)
```
