    '> Simple Balanced Parentheses Checking Using Stack</a> </h1>

```python
def parChecker(symbolString):
    s = Stack()
    balanced = True
    index = 0
    while index < len(symbolString) and balanced:
        symbol = symbolString[index]
        if symbol == "(":
            s.push(symbol)
        else:
            if s.isEmpty():
                balanced = False
            else:
                s.pop()

        index = index + 1

    if balanced and s.isEmpty():
        return True
    else:
        return False

def test ():
    A = ["(()()()())", "(((())))", "(()((())()))"]
    B = ["((((((())", "()))", "(()()(()"]
    for x in A:
        assert parChecker ( x )
    for x in B:
        assert not parChecker ( x )
    print ("done! ok!")
test()
```

<h1> 3.8.  <a href='
http://interactivepython.org/runestone/static/pythonds/BasicDS/ConvertingDecimalNumberstoBinaryNumbers.html
'> 
Converting Decimal Numbers to Binary Numbers Using Stack </a> </h1>

```python
from pythonds.basic.stack import Stack

def baseConverter(decNumber,base):
    digits = "0123456789ABCDEF"

    remstack = Stack()

    while decNumber > 0:
        rem = decNumber % base
        remstack.push(rem)
        decNumber = decNumber // base

    newString = ""
    while not remstack.isEmpty():
        newString = newString + digits[remstack.pop()]

    return newString

assert baseConverter(25,2) == bin(25)[2:]
assert baseConverter(25,16) == hex (25)[2:]
```

<h3> Infix-to-Postfix Conversion Algorithm  </h3>

Create an empty stack called opstack for keeping operators. 
Create an empty list for output.
Convert the input infix string to a list by using the string method split.
Scan the token list from left to right.

* If the token is an operand, append it to the end of the output list.
* If the token is a left parenthesis, push it on the opstack.
* If the token is a right parenthesis, 
  pop the opstack until the corresponding left parenthesis is removed. 
  Append each operator to the end of the output list.
  
* If the token is an operator, *, /, +, or -, push it on the opstack. 
However, first remove any operators already on the opstack that have higher or equal precedence and append them to the output list.
* When the input expression has been completely processed, check the opstack
  . Any operators still on the stack can be removed and appended to the end of the output list.

```python
class Stack:
    def __init__(self):
        self.items = []
        
    def tolist (self):
        return self.items
    
    def __str__(self):
        return str ( self.items)

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)
```

```python
#!/usr/bin/env python
import string
def tokenizer (infixexpr ):   # xxx allow no spaces in the expresson
    s= infixexpr[0]
    n = len (infixexpr )
    for i in range (1, n ) :
        op = "()+-*/" 
        x = infixexpr[i-1]
        y = infixexpr[i]
        if  (x+y!="//") and ( (x in op) or  ( y in op) ):        
            s += " "
        s+= infixexpr[i]
    return s.split ()

def traceHeader (infixexpr):
        s =  "\nTrace infixToPostfix:" + infixexpr + "\n"
        s += "{0:^4}{1:^7}{2:^5}{3:^10}".format ( "step", "infix", "stack", "postfix" ) 
        return (s)

def trace (cnt,input, output, stack ):
        x = "".join ( stack.tolist() )
        s = "{0:^4}{1:^7}{2:^5}{3:^10}".format(cnt, input, x[::-1], output)
        return (s)

def displayHorizontal (L, display):
    if  not display:
        return 
    for x in L:
        print(x)
    # return 
    n = len(L[0])
    for i in (range (n)) :     # print column i
        for s in  (L):
            print (s[i], end=" ")
        print ()
def infixToPostfix(infixexpr, display=False):
    prec = {'*': 3, '/': 3 , '//': 3, '+':2 ,'-': 2, '(': 1 }

    opStack = Stack()
    postfixList = []       
    tokenList = tokenizer (infixexpr)
    cnt = 0
    s = traceHeader (infixexpr ); 
    print (s)
    L = []
    for token in tokenList:
        if token.isalnum() :
            postfixList.append(token)
            s =  trace (cnt,token, token, opStack); cnt += 1; L += [s]
        elif token == '(':
            opStack.push(token)
            s = trace (cnt,token, "" , opStack); cnt += 1; L += [s]

        elif token == ')':
            s =  trace (cnt,token, "" , opStack); cnt += 1; L += [s]
            topToken = opStack.pop()
            while topToken != '(':
                s = trace (cnt, "", topToken , opStack); cnt += 1; L += [s]
                postfixList.append(topToken)
                topToken = opStack.pop()
        elif token in "+-*//" :
            topToken = ""
            w=token
            while  (not opStack.isEmpty()) and \
                (prec[opStack.peek()] >= prec[token]):
                s=  trace (cnt, w, topToken , opStack);  cnt += 1; L += [s]
                w=""
                topToken = opStack.pop() 
                postfixList.append( topToken )
            opStack.push(token)
            s= trace (cnt, w,  topToken, opStack);  cnt += 1; L += [s]

        else:
          raise ValueError ('Unrecognized token ' + token + "!")

    while not opStack.isEmpty():
        w = opStack.peek()
        s = trace (cnt, "", w , opStack); cnt += 1; L += [s]
        topToken = opStack.pop()
        postfixList.append( topToken )
    
    displayHorizontal (L, display)
    return " ".join(postfixList)

        
def tracePostfixEval (token, stack):
    x = ""
    for s in stack.tolist():
        x = " " + str (s) + x
    return "{0:^15}{1:>20}".format(token, x.strip()  )        
    
def postfixEval(postfixExpr, display=False):
    operandStack = Stack()
    tokenList = postfixExpr.split()
    s = "Trace Postfix Expression Evaluation: " + postfixExpr;
    print (s)
    s = "{0:^15}{1:>20}".format ("token ", "operand Stack" )
    L = [s]

    for token in tokenList:
        if token.isdigit() :         # operand
            operandStack.push(int(token))
            s = tracePostfixEval (token,operandStack );  L += [s]
        else:                             # operator
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token,operand1,operand2)
            operandStack.push(result)
            s = tracePostfixEval ( token, operandStack ) ;  L += [s]
    if display:
        for s in L:
            print (s)
    return operandStack.pop()

def doMath(op, op1, op2): 
    if op == "*":
        return op1 * op2
    elif op == "//":
        return op1 // op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    elif op == "-":
        return op1 - op2
    else: 
      raise RuntimeError ('Unrecognized token ' + token + "!")

def infixEval(s, display=False): # xxx HW 3.3
    return postfixEval ( infixToPostfix(s, display ), display )

#main()
display = False
postfixEval ( "33 2 * 2 +", True )
infixEval ("11+2-33//3", display)

L = [  "11+2-33//3", "A*B+C*D", ]
for s in L:
    infixToPostfix (s, True)
```

## Deque: Palindrome Checker

```python
#!/usr/bin/env python
"""
3.12 Extend the program from Listing 2.15 to handle palindromes with spaces. 
For example, I PREFER PI is a palindrome that reads the same forward and backward if you ignore the blank characters.
"""
from pythonds.basic.deque import Deque
import string

def palchecker(aString):
    aString = ''.join(e for e in aString if e.isalpha())
    aString = aString.upper()
    chardeque = Deque()

    for ch in aString:
        chardeque.addRear(ch)

    stillEqual = True

    while chardeque.size() > 1 and stillEqual:
        first = chardeque.removeFront()
        last = chardeque.removeRear()
        if first != last:
            stillEqual = False

    return stillEqual
A = [ "A Nut for a Jar of Tuna", "A But Tuba", "As I Pee, Sir, I See Pisa", \
        "A Toyota's a Toyota", "No X in Nixon", "radar",  "I prefer pi", "Race car" ] 
for s in A:
    assert palchecker (s)
assert not (palchecker("lsdkjfskf"))
```

* The linked list implementation given above is called a singly linked list because each node has a single reference to the next node in sequence. 
* An alternative implementation is known as a doubly linked list. 
* In this implementation, each node has a reference to the next node (commonly called next) as well as a reference to the preceding node (commonly called back). 
* The head reference also contains two references, one to the first node in the linked list 
and one to the last. 
* Code this implementation in Python.

```python
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = self.previous = None

    def getData(self):
        return self.data

    def __str__(self):
        return str(self.data)

    def getNext(self):
        return self.next
    
    def getPrevious(self):
        return self.previous
    
    def setData(self,newdata):
        self.data = newdata

    def setNext(self,newnext):
        self.next = newnext
        
    def setPrevious (self, p):
        self.previous = p
        

class doubleLinkedList:

    def __init__(self):
        self.count = 0
        self.head = None

    def isEmpty(self):
        return self.count == 0
            
    def insert (self,item, curr ):      # xxx add at the list head
        self.count += 1
        temp = Node(item)
        if self.head ==None:
            temp.setNext( temp )
            temp.setPrevious (temp)
            self.head = temp
        else:
            prev  = curr.getPrevious ()
            temp.setNext( curr )
            temp.setPrevious ( prev )
            curr.setPrevious ( temp )
            prev.setNext ( temp )
        return temp
    
    def addFront (self, item):
        self.head = self.insert ( item, self.head )
        
    def addRear (self, item):
        self.insert ( item, self.head )
        
    def add (self,item):      # xxx add at the list head
        self.addFront (item)
  
    def __getitem__(self,i):
        cnt = 0 
        current = self.head
        while current != None and cnt < i :
            current = current.getNext()
            cnt += 1
        if ( current==None):
          return None
        else:
          return current.getData()
 
    def toList (self):
        ans = []
        curr = self.head
        while curr != None:
            ans.append( curr.getData() )
            curr = curr.getNext()
            if curr==self.head:
                break;
        return  ans
    
    def __str__(self):
        return str ( self.toList() )
        
    def length(self): 
        return self.count

    def find (self,item):
        curr = self.head
        found = False
        while curr != None:
            if curr.getData() == item:
                return curr
            curr = curr.getNext()
            if curr==self.head:
                break;
        return None 
    
    def search(self,item):
        return self.find ( item ) != None

    def remove (self,item):
        curr = self.find ( item ) 
        if curr == None:
            print ( "Cannot remove %d" % (item) )
        else:
            self.removeNode  ( curr )
        
    def removeNode (self,curr ):    
        self.count -= 1
        if self.count==0:
            self.head = None
        prev = curr.getPrevious()
        next = curr.getNext()
        prev.setNext( next )
        next.setPrevious( prev )
        if curr == self.head :
            self.head = next
        return curr.getData()
            
    def removeFront ( self):
        x = self.head
        if x != None:
            x = self.removeNode ( x )
        return x

    def removeRear ( self):
        x = self.head
        if x != None:
            x = x.getPrevious()
            x = self.removeNode ( x )
        return x

def testDoubleLinkedList (n):          
    mylist = doubleLinkedList()
    for i in range (n):
        mylist.add(i)
    for i in range (n+1):
        mylist.remove (i)
    return  mylist.length() 


def testDoubleLinkedList2 (): 
    mylist = doubleLinkedList()
    for i in [31,77,17,93,26,54]:
        mylist.add(i)

    assert mylist.toList() == ([54, 26, 93, 17, 77, 31])  
    assert mylist.length() ==6 
    assert mylist.search(93)
    assert not mylist.search(100)


    mylist.add(100)
    assert mylist.search(93)

    mylist.remove(54)
    
    mylist.remove(93)
    mylist.remove(31)
    mylist.remove(93)
    assert str(mylist)==str ([100, 26, 17, 77]) 
    assert not mylist.search(93)
        
def testDoubleLinkedList3 (): 
    mylist = doubleLinkedList()
    for i in [31,77,17,93,26,54]:
        mylist.add(i)
    mylist.removeFront ( )
    mylist.removeRear()
    assert mylist.toList()==[26, 93, 17, 77]  
    assert mylist.search(93)
    mylist.removeFront ( )
    mylist.removeRear()
    assert mylist.toList() == [93, 17]
    mylist.removeFront ( )
    mylist.removeRear()
    assert mylist.toList() == []
testDoubleLinkedList2()
testDoubleLinkedList3()
```

<h3>
  Palindrome: an  application of  dequeue
</h3>

```python
import string
def palchecker(s):
    exclude = set(string.punctuation)
    exclude.add ( " ")
    s = ''.join(ch for ch in s if ch not in exclude)
    aString = s.upper() 
    chardeque = newDeque()

    for ch in aString:
        chardeque.addRear(ch)

    stillEqual = True

    while chardeque.size() > 1 and stillEqual:
        first = chardeque.removeFront()
        last = chardeque.removeRear()
        if first != last:
            stillEqual = False

    return stillEqual

assert not palchecker("lsdkjfskf")
assert palchecker("radar")
assert palchecker("racecar")
assert palchecker("Anne, I vote more cars race Rome to Vienna")
```
