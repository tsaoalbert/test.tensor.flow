<h1>   <a href='http://interactivepython.org/runestone/static/pythonds/BasicDS/TheStackAbstractDataType.html    '>  Stack Abstract Data Type </a> </h1>
* 


* Stack() creates a new stack that is empty. It needs no parameters and returns an empty stack.
* push(item) adds a new item to the top of the stack. It needs the item and returns nothing.
* pop() removes the top item from the stack. It needs no parameters and returns the item. The stack is modified.
* peek() returns the top item from the stack but does not remove it. It needs no parameters. The stack is not modified.
* isEmpty() tests to see whether the stack is empty. It needs no parameters and returns a boolean value.
size() returns the number of items on the stack. It needs no parameters and returns an integer.

```python
class Stack:
    def __init__(self):
        self.items = []              # implementation based on Python list

    def __str__ (self):
        return str (self.items)
    
    
    def isEmpty(self):
        return self.items == []

    def push(self, x):
        self.items.append(x)              # Add at the end of the list  O(1)

    def pop(self):
        return self.items.pop()           # remove at the end of the list  O(1)

    def peek(self):
        return self.items[-1]

    def size(self):
        return len(self.items)

```

<h3>
Queue implementation based on Python list
</h3>

```python
#!/usr/bin/env python
class Queue:
    def __init__(self):
        self.items = []              # implementation based on Python list

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):         # add at the front of the list O(N)
        self.items.insert(0,item)

    def dequeue(self):               # remove at the end of the list  O(1)
        return self.items.pop()

    def size(self):
        return len(self.items)
```

<h3> Deque implementation based on Python list </h3>
<h4>
 <ul>
     <li> Note that Python collection also has built-in container  <a href=" https://docs.python.org/2/library/collections.html "> 
 deque    </a>
     </li>
    </ul>
</h4>

```python

class Deque:                            # Deque implementation based on python list
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []
    # O(N)                              # Equivalent Python deque methods
    def addFront(self, item):           # appendLeft
        self.items.insert(0,item)

    # O(1)
    def addRear(self, item):            # append
        self.items.append(item)
    # O(N)
    def removeFront(self):              # popLeft
        return self.items.pop(0)
    # O(1)
    def removeRear(self):               # pop
        return self.items.pop()

    def size(self):
        return len(self.items)
```

```python
def toString (self):
    return str (self.items)
 
Deque.__str__ = toString
Deque.__len__ = Deque.size
```

<h3> Palindrome Checker With Deque </h3>

```python
#!/usr/bin/env python
"""
3.12 Extend the program from Listing 2.15 to handle palindromes with spaces. 
For example, I PREFER PI is a palindrome that reads the same forward and backward if you ignore the blank characters.
"""
from pythonds.basic.deque import Deque
import string

def palchecker(aString):
    aString = ''.join(e for e in aString if e.isalpha())
    aString = aString.upper()
    chardeque = Deque()

    for ch in aString:
        chardeque.addRear(ch)

    stillEqual = True

    while chardeque.size() > 1 and stillEqual:
        first = chardeque.removeFront()
        last = chardeque.removeRear()
        if first != last:
            stillEqual = False

    return stillEqual
A = [ "A Nut for a Jar of Tuna", "A But Tuba", "As I Pee, Sir, I See Pisa", \
        "A Toyota's a Toyota", "No X in Nixon", "radar",  "I prefer pi", "Race car" ] 
for s in A:
    assert palchecker (s)
assert not (palchecker("lsdkjfskf"))
```

    '> Simple Balanced Parentheses Checking Using Stack</a> </h1>

```python
def parChecker(symbolString):
    s = Stack()
    balanced = True
    index = 0
    while index < len(symbolString) and balanced:
        symbol = symbolString[index]
        if symbol == "(":
            s.push(symbol)
        else:
            if s.isEmpty():
                balanced = False
            else:
                s.pop()

        index = index + 1

    if balanced and s.isEmpty():
        return True
    else:
        return False

def test ():
    A = ["(()()()())", "(((())))", "(()((())()))"]
    B = ["((((((())", "()))", "(()()(()"]
    for x in A:
        assert parChecker ( x )
    for x in B:
        assert not parChecker ( x )
    print ("done! ok!")
test()
```

```python
from pythonds.basic.stack import Stack

def baseConverter(decNumber,base):
    digits = "0123456789ABCDEF"

    remstack = Stack()

    while decNumber > 0:
        rem = decNumber % base
        remstack.push(rem)
        decNumber = decNumber // base

    newString = ""
    while not remstack.isEmpty():
        newString = newString + digits[remstack.pop()]

    return newString

assert baseConverter(25,2) == bin(25)[2:]
assert baseConverter(25,16) == hex (25)[2:]
```

<h3> Infix-to-Postfix Conversion Algorithm  </h3>

Create an empty stack called opstack for keeping operators. 
Create an empty list for output.
Convert the input infix string to a list by using the string method split.
Scan the token list from left to right.

* A. If the token is an operand, append it to the end of the output list.
* B. If the token is a (, push it on the opstack.
* C. If the token is a ), pop the opstack until the corresponding ( is removed. Append each operator to the end of the output list.
* D. If the token is an arithmatic operator, *, /, +, or -, push it on the opstack. 
* E. However, first remove any operators already on the opstack that have higher or equal precedence and append them to the output list.
* F. When the input expression has been completely processed, append any operators still on the stack  to the end of the output list.

```python
class Stack:
    def __init__(self):
        self.items = []
        
    def tolist (self):
        return self.items
    
    def __str__(self):
        return str ( self.items)

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)
```

```python
#!/usr/bin/env python
import string
def tokenizer (infixexpr ):   # xxx allow no spaces in the expresson
    s= infixexpr[0]
    n = len (infixexpr )
    for i in range (1, n ) :
        op = "()+-*/" 
        x = infixexpr[i-1]
        y = infixexpr[i]
        if  (x+y!="//") and ( (x in op) or  ( y in op) ):        
            s += " "
        s+= infixexpr[i]
    return s.split ()

def traceHeader (infixexpr, display):
    if display:
        s =  "\nTrace infixToPostfix:" + infixexpr + "\n"
        s += "{0:^4}{1:^6}{2:^7}{3:^5}{4:^10}".format ( "step", "Rule", "infix", "stack", "postfix" ) 
        print (s)


def displayHorizontal (L, display):
    if  not display:
        return 
    index = 3

    msg = [ "Step", "Rule", "Infix", ["Stack"], "Postfix"]
    n = 0
    for x in L:
        n = max (n, len(x[index]))
    M = []
    L = [msg] + L
    for x in L:
        z = [x[0],x[1], x[2]]
        m= n - len( x[index] )
        for i in range (m):
            z.append( " ")
        for i in reversed( x[index ] ):
            z.append(i)
        z.append ( x[-1])
        M.append (z)
    n = len(M[0])
    for i in (range (n)) :     # print row i
        s =  ""
        for x in  M:
            s += "{0:8}".format ( x[i] )
        print (s)
        if   i <=2 or i ==n-2 :
            t = "-"* len(s)
            print ( t )
        
def updateList (L, step, token, opStack, y):
    x = opStack.tolist().copy()

    cnt = 0
    if L!=[]:
        cnt = L[-1][0]+1  
    s = chr ( ord('A') -1 + step )
    z = [cnt, s, token, x, y ]
    L.append ( z )
    
def infixToPostfix(infixexpr, display=False):
    prec = {'*': 3, '/': 3 , '//': 3, '+':2 ,'-': 2, '(': 1 }

    opStack = Stack()
    postfixList = []       
    tokenList = tokenizer (infixexpr)
    cnt = 0
    traceHeader (infixexpr, display ); 
    L = []
    """
    A. If the token is an operand, append it to the end of the output list.
B. If the token is a (, push it on the opstack.
C. If the token is a ), pop the opstack until the corresponding ( is removed. 
D. If the token is an operator, push it on the opstack. 
D1. However, first remove any operators already on the opstack that have higher or equal precedence and append them to the output list.
E. When the input expression has been completely processed, append any operators still on the stack  to the end of the output list.
~ 
    """
    for token in tokenList:
        if token.isalnum() :   # Step A.
            postfixList.append(token)
            updateList (L, 1, token, opStack, token)
        elif token == '(':   # Step B
            opStack.push(token)
            updateList (L, 2, token, opStack, "")
        elif token == ')':    #Step C
            updateList (L, 3, token, opStack, "")
            topToken = opStack.pop()
            while topToken != '(':
                updateList (L, 3, "", opStack, topToken)
                postfixList.append(topToken)
                topToken = opStack.pop()
        elif token in "+-*//" : #Step D
            topToken = ""
            w=token
            while  (not opStack.isEmpty()) and \
                (prec[opStack.peek()] >= prec[token]): # Step E
                updateList (L, 4, w, opStack, topToken)
                w=""
                topToken = opStack.pop() 
                postfixList.append( topToken )
            opStack.push(token)
            updateList (L, 5, w, opStack, topToken)
        else:
          raise ValueError ('Unrecognized token ' + token + "!")

    while not opStack.isEmpty(): # Step E
        w = opStack.peek()
        updateList (L, 6, "", opStack, w)
        topToken = opStack.pop()
        postfixList.append( topToken )
    
    displayHorizontal (L, display)
    return " ".join(postfixList)

def foo (L ):
    for x in L:
        print ("{0:^15}".format(x[0]), end="")
        for y in x[1:][::-1]:
            print ("{0:>3}".format(y), end="")
        print (  )
def tracePostfixEval (token, stack):
    L = [token] +  stack.tolist().copy()
    return L
        
def displayPostfixEval (L):
    m = 0
    for x in L:
        m = max (m, len(x))
    M = []
    s = ["-"*10]
    for x in L:
        k = m -len(x)
        for i in range ( k ):
            x.append ("") 
        M.append ( s+ x[0:1] + s + x[1:][::-1] +s   )  
    L = M 
    for j in range (  len(L[0]) ):
        for i in range ( len(L) ):
            print ("{0:>10}".format (L[i][j]), end="")
        print ()
    print ()

        
def postfixEval(postfixExpr, display=False):
    operandStack = Stack()
    tokenList = postfixExpr.split()
    if display:
        s = "Trace Postfix Expression Evaluation: " + postfixExpr;
        print (s)
    L = [["token", "stack", "operand",  ]]

    for token in tokenList:
        if token.isdigit() :         # operand
            operandStack.push(int(token))
            s = tracePostfixEval (token,operandStack );  L += [s]
        else:                             # operator
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token,operand1,operand2)
            operandStack.push(result)
            s = tracePostfixEval ( token, operandStack ) ;  L += [s]
    if display:
        displayPostfixEval(L)

    return operandStack.pop()

def doMath(op, op1, op2): 
    if op == "*":
        return op1 * op2
    elif op == "//":
        return op1 // op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    elif op == "-":
        return op1 - op2
    else: 
      raise RuntimeError ('Unrecognized token ' + token + "!")

def infixEval(s, display=False): # xxx HW 3.3
    return postfixEval ( infixToPostfix(s, display ), display )

#main()
display = False

assert 68 == postfixEval ( "33 2 * 2 +", display )
assert 2 == infixEval ("11+2-33//3", display)
print ("pass!!!")
```

<h1> <font color="red">
    Doubly Linked Lists </font>

</h1>

* The linked list implementation given above is called a singly linked list because each node has a single reference to the next node in sequence. 
* An alternative implementation is known as a doubly linked list. 
* In this implementation, each node has a reference to the next node (commonly called next) as well as a reference to the preceding node (commonly called back). 
* The head reference also contains two references, one to the first node in the linked list 
and one to the last.

```python
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = self.previous = None

    def getData(self):
        return self.data

    def __str__(self):
        return str(self.data)

    def getNext(self):
        return self.next
    
    def getPrevious(self):
        return self.previous
    
    def setData(self,newdata):
        self.data = newdata

    def setNext(self,newnext):
        self.next = newnext
        
    def setPrevious (self, p):
        self.previous = p
        

class doublyLinkedList:

    def __init__(self):
        self.count = 0
        self.head = None

    def isEmpty(self):
        return self.count == 0
            
    def insert (self,item, curr ):      # xxx add at the list head
        self.count += 1
        temp = Node(item)
        if self.head ==None:
            temp.setNext( temp )
            temp.setPrevious (temp)
            self.head = temp
        else:
            prev  = curr.getPrevious ()
            temp.setNext( curr )
            temp.setPrevious ( prev )
            curr.setPrevious ( temp )
            prev.setNext ( temp )
        return temp
    
    def addFront (self, item):
        self.head = self.insert ( item, self.head )
        
    def addRear (self, item):
        self.insert ( item, self.head )
        
    def add (self,item):      # xxx add at the list head
        self.addFront (item)
  
    def __getitem__(self,i):
        cnt = 0 
        current = self.head
        while current != None and cnt < i :
            current = current.getNext()
            cnt += 1
        if ( current==None):
          return None
        else:
          return current.getData()
 
    def toList (self):
        ans = []
        curr = self.head
        while curr != None:
            ans.append( curr.getData() )
            curr = curr.getNext()
            if curr==self.head:
                break;
        return  ans
    
    def __str__(self):
        return str ( self.toList() )
        
    def length(self): 
        return self.count

    def find (self,item):
        curr = self.head
        found = False
        while curr != None:
            if curr.getData() == item:
                return curr
            curr = curr.getNext()
            if curr==self.head:
                break;
        return None 
    
    def search(self,item):
        return self.find ( item ) != None

    def remove (self,item):
        curr = self.find ( item ) 
        if curr == None:
            print ( "Cannot remove %d" % (item) )
        else:
            self.removeNode  ( curr )
        
    def removeNode (self,curr ):    
        self.count -= 1
        if self.count==0:
            self.head = None
        prev = curr.getPrevious()
        next = curr.getNext()
        prev.setNext( next )
        next.setPrevious( prev )
        if curr == self.head :
            self.head = next
        return curr.getData()
            
    def removeFront ( self):
        x = self.head
        if x != None:
            x = self.removeNode ( x )
        return x

    def removeRear ( self):
        x = self.head
        if x != None:
            x = x.getPrevious()
            x = self.removeNode ( x )
        return x

def testDoublyLinkedList (n):          
    mylist = doublyLinkedList()
    for i in range (n):
        mylist.add(i)
    for i in range (n+1):
        mylist.remove (i)
    return  mylist.length() 


def testDoublyLinkedList2 (): 
    mylist = doublyLinkedList()
    for i in [31,77,17,93,26,54]:
        mylist.add(i)

    assert mylist.toList() == ([54, 26, 93, 17, 77, 31])  
    assert mylist.length() ==6 
    assert mylist.search(93)
    assert not mylist.search(100)


    mylist.add(100)
    assert mylist.search(93)

    mylist.remove(54)
    
    mylist.remove(93)
    mylist.remove(31)
    mylist.remove(93)
    assert str(mylist)==str ([100, 26, 17, 77]) 
    assert not mylist.search(93)
        
def testDoublyLinkedList3 (): 
    mylist = doublyLinkedList()
    for i in [31,77,17,93,26,54]:
        mylist.add(i)
    mylist.removeFront ( )
    mylist.removeRear()
    assert mylist.toList()==[26, 93, 17, 77]  
    assert mylist.search(93)
    mylist.removeFront ( )
    mylist.removeRear()
    assert mylist.toList() == [93, 17]
    mylist.removeFront ( )
    mylist.removeRear()
    assert mylist.toList() == []
testDoublyLinkedList2()
testDoublyLinkedList3()
```

<h3>
  Palindrome: an  application of  dequeue
</h3>

```python
import string
def palchecker(s):
    exclude = set(string.punctuation)
    exclude.add ( " ")
    s = ''.join(ch for ch in s if ch not in exclude)
    aString = s.upper() 
    chardeque = newDeque()

    for ch in aString:
        chardeque.addRear(ch)

    stillEqual = True

    while chardeque.size() > 1 and stillEqual:
        first = chardeque.removeFront()
        last = chardeque.removeRear()
        if first != last:
            stillEqual = False

    return stillEqual

assert not palchecker("lsdkjfskf")
assert palchecker("radar")
assert palchecker("racecar")
assert palchecker("Anne, I vote more cars race Rome to Vienna")
```
